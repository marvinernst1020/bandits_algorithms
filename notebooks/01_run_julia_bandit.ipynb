# Turing.jl Bandit Simulation: External Run

This notebook runs the `turing_bandit_setup.jl` script using a shell command, monitors execution, and loads the output plot into the notebook.


**This is what we need to run in the terminal:**

#### julia turing_bandit_setup.jl

```python
import subprocess

# Path to the Julia script
script_path = "../bandits_turing/turing_bandit_setup.jl"

# Run the script
julia_path = "/home/ice/.juliaup/bin/julia"  # Replace with your path! go to the terminal and put: which julia
result = subprocess.run([julia_path, script_path], capture_output=True, text=True)


# Print stdout/stderr to check for issues
print("--- STDOUT ---")
print(result.stdout)
print("--- STDERR ---")
print(result.stderr)
```
